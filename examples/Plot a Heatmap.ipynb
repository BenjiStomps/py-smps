# Visualizing SMPS Data using the py-smps library



```python
import seaborn as sns
import os
import matplotlib.pyplot as plt

import smps

%matplotlib inline


sns.set_context('notebook', font_scale=1.5)

datadir = "/Users/dh/Documents/GitHub/py-smps/py-smps/sample-data/"

f1 = "SMPS_Number.txt"
f2 = "r20160728_SMPS_diameter.txt"
```

```python
s = smps.io.load_smps_file(os.path.join(datadir, f2))

ax = s.heatmap()
```

```python
X = s.df.index.values
Y = s.midpoints
Z = s.histogram.T.values

ax = smps.plots.heatmap(X, Y, Z)
```
